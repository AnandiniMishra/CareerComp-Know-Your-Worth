```python
import streamlit as st
import pandas as pd
import joblib
model = joblib.load("best_salary_model.pkl")
scaler = joblib.load("scaler.pkl")
model_features = joblib.load("model_features.pkl")
job_titles = joblib.load("job_titles.pkl")
st.markdown(
    """
    <style>
    body {
        background-color: #1e1e1e;
        color: #f5f5f5;
    }
    .stApp {
        background-color: #1e1e1e;
    }
    h1, h2, h3, h4 {
        color: #00b894;
    }
    .css-1cpxqw2 {
        color: #ffffff !important;
    }
    .css-1n76uvr {
        color: #000000 !important;
        background-color: #ffffff !important;
    }
    .stButton>button {
        color: white;
        background-color: #00b894;
        border-radius: 10px;
    }
    .stButton>button:hover {
        background-color: #009f74;
    }
    </style>
    """,
    unsafe_allow_html=True
)

# Title
st.markdown("<h1 style='text-align: center;'>ðŸ’¼ CareerCompâ„¢ â€“ Know Your Worth</h1>", unsafe_allow_html=True)
st.markdown("<p style='text-align: center; color: #bbb;'>Predict your future salary with data-driven intelligence</p>", unsafe_allow_html=True)
st.markdown("---")
age = st.slider("Select Age", 18, 70, 25)
experience = st.slider("Years of Experience", 0, 50, 2)
gender = st.selectbox("Gender", ["Male", "Female", "Other"])
education = st.selectbox("Education Level", ["High School", "Bachelor's", "Master's", "PhD"])
job_title = st.selectbox("Job Title", job_titles)
input_df = pd.DataFrame([{
    "Age": age,
    "YearsExperience": experience,
    "Gender": gender,
    "EducationLevel": education,
    "JobTitle": job_title
}])
input_processed = pd.get_dummies(input_df)
for col in model_features:
    if col not in input_processed.columns:
        input_processed[col] = 0
input_processed = input_processed[model_features]
input_scaled = scaler.transform(input_processed)
if st.button("ðŸ“Š Predict Salary"):
    predicted_salary = model.predict(input_scaled)[0]
    st.success(f"ðŸ’° Estimated Annual Salary: â‚¹{predicted_salary:,.2f}")
```
